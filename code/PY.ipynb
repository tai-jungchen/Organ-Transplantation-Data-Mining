#### Time on waitlist

```python
# helper function
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)
```

```python
lst = []
lst1 = []
lst2 = []
lst_std = []
lst1_std = []
lst2_std = []
for center in flist:
    
    Fullfname =  path + "/" + center
    df = pd.read_csv(Fullfname, index_col=0, encoding='utf-8', encoding_errors='ignore') 
    df2 = df.loc[(df['end_date'] > '2017-11-24')]
    df1 = df.loc[(df['end_date'] <= '2017-11-24')]
    
    ########## edit features below ##########
    df['time_on_wait_list'] = pd.to_datetime(df['end_date']) - pd.to_datetime(df['init_date'])
    df.loc[df["time_on_wait_list"] > pd.Timedelta("1654 days"), "time_on_wait_list"] = pd.Timedelta("1654 days")
    
    mean_wt = df['time_on_wait_list'].mean(skipna = True).days
    std_wt = df['time_on_wait_list'].std(ddof = 0).days
    lst.append(mean_wt)
    lst_std.append(std_wt)
    
    ########## edit features below ##########

centers_df['mean_time_on_waitlist'] = lst
# centers_df['mean_time_on_waitlist_p1'] = lst1
# centers_df['mean_time_on_waitlist_p2'] = lst2
centers_df['std_time_on_waitlist'] = lst_std
# centers_df['std_time_on_waitlist_p1'] = lst1_std
# centers_df['std_time_on_waitlist_p2'] = lst2_std
```

#### Patient years, waitlist mortality, and tx per 100-patient years

```python
# days_between('2020-03-01', '2020-03-02')
```

```python
pyear1_lst = []
pyear2_lst = []

wl_mor_p1_lst = []
wl_mor_p2_lst = []

tx_pyear_p1_lst = []
tx_pyear_p2_lst = []

tx_pyear_p1_local_lst = []
tx_pyear_p2_local_lst = []

tx_pyear_p1_regional_lst = []
tx_pyear_p2_regional_lst = []

tx_pyear_p1_national_lst = []
tx_pyear_p2_national_lst = []

for center in flist:
    
    Fullfname =  path + "/" + center
    df = pd.read_csv(Fullfname, index_col=0, encoding='utf-8', encoding_errors='ignore')
    p1_days_center = 0
    p2_days_center = 0
    
    death_cases_p1 = df.loc[(df['rem_cd']== 8) & (df['end_date'] <= '2017-11-24')]
    death_cases_p2 = df.loc[(df['rem_cd'] == 8) & (df['end_date'] > '2017-11-24')]
    
    num_death_p1 = len(death_cases_p1)
    num_death_p2 = len(death_cases_p2)
    
    tx_cases_p1 = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] <= '2017-11-24')]
    tx_cases_p2 = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] > '2017-11-24')]
    
    tx_cases_p1_local = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] <= '2017-11-24')& (df['share_ty'] == 3)]
    tx_cases_p2_local = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] > '2017-11-24') & (df['share_ty'] == 3)]
    
    tx_cases_p1_regional = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] <= '2017-11-24')& (df['share_ty'] == 4)]
    tx_cases_p2_regional = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] > '2017-11-24') & (df['share_ty'] == 4)]
    
    tx_cases_p1_national = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] <= '2017-11-24')& (df['share_ty'] == 5)]
    tx_cases_p2_national = df.loc[((df['rem_cd'] == 4) | (df['rem_cd'] == 21)) & (df['end_date'] > '2017-11-24') & (df['share_ty'] == 5)]
    
    num_of_tx_p1 = len(tx_cases_p1)
    num_of_tx_p2 = len(tx_cases_p2)
    
    num_of_tx_p1_local = len(tx_cases_p1_local)
    num_of_tx_p2_local = len(tx_cases_p2_local)
    
    num_of_tx_p1_regional = len(tx_cases_p1_regional)
    num_of_tx_p2_regional = len(tx_cases_p2_regional)
    
    num_of_tx_p1_national = len(tx_cases_p1_national)
    num_of_tx_p2_national = len(tx_cases_p2_national)
    
    ######### edit features below ##########
    for index, row in df.iterrows():
        end_date = row['end_date']
        init_date = row['init_date']
        p1_days = 0
        p2_days = 0
        
        # patient 1
        if (end_date <= '2017-11-24') & (init_date >= '2015-08-20'):
            p1_days = days_between(end_date, max(init_date, '2015-08-20'))
        
        # patient 0
        elif (end_date <= '2017-11-24') & (init_date < '2015-08-20'):
            p1_days = days_between(end_date, max(init_date, '2015-08-20'))
        
        # patient 3, 5, 6, 7
        elif (end_date > '2017-11-24') & (init_date <= '2017-11-24'):
            p1_days = days_between('2017-11-24', max(init_date, '2015-08-20'))
            p2_days = days_between(min(end_date, '2020-03-01'), '2017-11-25')
        
        # patient 2 and 4
        elif (end_date > '2017-11-24') & (init_date > '2017-11-24'):
            p2_days = days_between(min(end_date, '2020-03-01'), init_date)
        
        else:    # check exception
            print('error')
            print(f'init_date: {init_date}')
            print(f'end_date: {end_date}')
            print()
            
        # cumulate the days per center
        p1_days_center += p1_days
        p2_days_center += p2_days
    
    # calculate center's patient years
    pyear1 = p1_days_center / 365.25
    pyear2 = p2_days_center / 365.25
    
    # calculate wait list mortality per center
    if pyear1 == 0:
        wl_mor_p1 = 0
        tx_pyear_p1 = 0
        tx_pyear_p1_local = 0
        tx_pyear_p1_regional = 0
        tx_pyear_p1_national = 0
        
    elif pyear2 == 0:
        wl_mor_p2 = 0
        tx_pyear_p2 = 0
        tx_pyear_p2_local = 0
        tx_pyear_p2_regional = 0
        tx_pyear_p2_national = 0
        
    else:
        wl_mor_p1 = num_death_p1 / pyear1 * 100    # per 100 patient years
        wl_mor_p2 = num_death_p2 /pyear2 * 100
        
        tx_pyear_p1 = num_of_tx_p1 / pyear1 * 100    
        tx_pyear_p2 = num_of_tx_p2 /pyear2 * 100
        
        tx_pyear_p1_local = num_of_tx_p1_local / pyear1 * 100
        tx_pyear_p2_local = num_of_tx_p2_local /pyear2 * 100
        
        tx_pyear_p1_regional = num_of_tx_p1_regional / pyear1 * 100    
        tx_pyear_p2_regional = num_of_tx_p2_regional /pyear2 * 100
        
        tx_pyear_p1_national = num_of_tx_p1_national / pyear1 * 100
        tx_pyear_p2_national = num_of_tx_p2_national /pyear2 * 100
    
    pyear1_lst.append(pyear1)
    pyear2_lst.append(pyear2)
    wl_mor_p1_lst.append(wl_mor_p1)
    wl_mor_p2_lst.append(wl_mor_p2)
    tx_pyear_p1_lst.append(tx_pyear_p1)
    tx_pyear_p2_lst.append(tx_pyear_p2)
    tx_pyear_p1_local_lst.append(tx_pyear_p1_local)
    tx_pyear_p2_local_lst.append(tx_pyear_p2_local)
    tx_pyear_p1_regional_lst.append(tx_pyear_p1_regional)
    tx_pyear_p2_regional_lst.append(tx_pyear_p2_regional)
    tx_pyear_p1_national_lst.append(tx_pyear_p1_national)
    tx_pyear_p2_national_lst.append(tx_pyear_p2_national)
    ######### edit features below ##########

centers_df['patient_year_p1'] = pyear1_lst
centers_df['patient_year_p2'] = pyear2_lst
centers_df['wl_mortality_p1(per_100-patient_years)'] = wl_mor_p1_lst
centers_df['wl_mortality_p2(per_100-patient_years)'] = wl_mor_p2_lst
centers_df['tx_p1(per_100-patient_years)'] = tx_pyear_p1_lst
centers_df['tx_p2(per_100-patient_years)'] = tx_pyear_p2_lst
centers_df['tx_p1_local(per_100-patient_years)'] = tx_pyear_p1_local_lst
centers_df['tx_p2_local(per_100-patient_years)'] = tx_pyear_p2_local_lst
centers_df['tx_p1_regional(per_100-patient_years)'] = tx_pyear_p1_regional_lst
centers_df['tx_p2_regional(per_100-patient_years)'] = tx_pyear_p2_regional_lst
centers_df['tx_p1_national(per_100-patient_years)'] = tx_pyear_p1_national_lst
centers_df['tx_p2_national(per_100-patient_years)'] = tx_pyear_p2_national_lst
```
